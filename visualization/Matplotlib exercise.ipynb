```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
# Exercise: 1
# 1. Create afigure object
# 2. Use add_axes to add an axes to the figurecanvas [0,0,1,1]
# 3. Plot(x,y) on that axes and set the labels and titles
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')
```
