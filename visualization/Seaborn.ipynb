```python
# Distribution plots on seaborn
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# predefined data
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
# Dist plot on a one colums
# KDE: Kernal destination 
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
# Joint plot
sns.jointplot(x='total_bill', y='tip', data=tips,kind='hex')
```

```python
# Joint plot
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# Joint plot
# create regression line
sns.jointplot(x='total_bill', y='tip', data=tips,kind='reg')
```

```python
# Joint plot
sns.jointplot(x='total_bill', y='tip', data=tips,kind='kde')
```

```python
# pair plot
# for all columns
# quickly visualize your data
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
# Categorical plots
```

```python
tips
```

```python
# distribution of particular category such as tip, time, day etc
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
# it shows the average
```

```python
import numpy as np
# get std estimator
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
# count plot, count the occurences so there is no y axis
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights.head()
```

```python
# Heat maps, for matrix plots
# Data should always in matrix form
```

```python
# corelation data
tc = tips.corr()
tc
```

```python
# heat map
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
flights
```

```python
# create pivot tables
f = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
flights
```

```python
f
```

```python
sns.heatmap(f, cmap='magma', linecolor='white', lw=4)
```

```python
# cluster map, use hierarchical clusterting
```

```python
sns.clustermap(f, cmap='coolwarm', standard_scale=1)
# cluster based on similarity
```
