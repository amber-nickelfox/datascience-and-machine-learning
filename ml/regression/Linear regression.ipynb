```python
# Single linear regression with one independent and feature variable
```

```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# iloc, locate indexes
# Problem statement
# salary dependent variable to predict
# In dependent variable is year experience
# this dataset consist only 2 columns and build simple linear
# regression model
# x = experience
# y = salary
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
# create 4 seperate set
# 1. X train: matrix of feature of training set
# 2. X test: matrix of feature of test set
# 3. Y train: matrix of independent variable of train set 
# 4. Y Test: matrix of independent variable of test set
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 1)
```

```python
# Training part: Understand the relation between the salary and experience
```

```python
# import simple linear regression model
# regression = continuity
# classification = category
from sklearn.linear_model import LinearRegression
# get a model by creating instance
regressor = LinearRegression()
# fit method to train model
regressor.fit(x_train, y_train)
```

```python
# predict the test set result
# ground truth if as expected
# we enter x test: rest of the experience and it will
# predict the y test: rest of the salary
# y_pred = predicted salary
y_pred = regressor.predict(x_test)
```

```python
# compare y_pred (predicted salary) and y_test (actual salary)
```

```python
# Visualize training set result
# below is x training (exp) and y training (salary)
plt.scatter(x_train, y_train, color='red')
# plot the regression line for prediction: should be as close as possible
# below is x training, x predict
plt.plot(x_train, regressor.predict(x_train))
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
```

```python
# Visualize test set result
plt.scatter(x_test, y_test, color='red')
# plot the regression line for prediction: should be as close as possible
# below is x training, x predict
plt.plot(x_test, y_pred)
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
```
