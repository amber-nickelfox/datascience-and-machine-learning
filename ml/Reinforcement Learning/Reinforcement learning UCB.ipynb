##### Reinforcement Learning: Reinforcement Learning is a powerful branch of Machine Learning. It is used to solve interacting problems where the data observed up to time t is considered to decide which action to take at time t + 1. It is also used for Artificial Intelligence when training machines to perform tasks such as walking. Desired outcomes provide the AI with reward, undesired with punishment. Machines learn through trial and error.

* Reinforcement learning use for trained robots to walk. 
* When create a AI robot dog, implement a reinforcement learning for walk, say all action for walk to dog. if peform well then give reward else punishment. (Reward = 1, punishment = 0)
* So if actions are good and correct  to try to repeat more and more.
* So they can learn to walk and find a way to walk on their own.

##### The multi armed Bandit Problem:

* What is Multi armed bandit problem, it is a game machine. handle on right. pull the handle or lever to initiate the game. 
* Why they called bandit, because these machine are quickiest way to loose money in casino. 50 % chance to loose money 50 % chance to win.
* This one is one armed, so many machine like this called multi armed bandit. 
* So how to maximize the profit by these by playing many times. 
* Assumption is: Each machine has distribution. 
* So goal is to figure out which one is best distribution for you.
* So when you know the distribution you will figure it out how to put money in which bandit.
* You need to explore the machine and same time exploit machine.
* Regret is not use optimal method
* At the end we need to find how to bet maximize your return.
* We agree behind every machine there is a certain distribution but we dont know, we need to combine explore these machine with explotation in order to find out which machine is best. 
* Then perform A/B test for find best.
