##### K nearest neighbour (K-NN)
1. Put a new data point in defined classifier
2. Example: put point x in category1 or category2 classifier.
3. KNN is used to identify to decide this.
4. Build a step by step guide

##### Steps for algorithms:
1. Choose the number k of neighbors (common or default value is 5)
2. Take the K neighbors to the new data point, according to Euclidean distance or also other distances (manhetten etc)
3. Among these K neighbors, count the number of data point in each category.
4. Assign the new data point to the category where you counted the most neignbors
5. Your model is ready
6. Now count the most nearest neighbour in both category and assign to them

```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# iloc, locate indexes
dataset = pd.read_csv('Social_Network_Ads.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
# create 4 seperate set
# 1. X train: matrix of feature of training set
# 2. X test: matrix of feature of test set
# 3. Y train: matrix of independent variable of train set 
# 4. Y Test: matrix of independent variable of test set
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 1)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Training the KNN model on training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, p=2, metric="minkowski")
classifier.fit(x_train, y_train)
```

```python
# predicting new result
classifier.predict(sc.transform([[30,87000]]))
```

```python
# predicting test result
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# confusion metrics
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
